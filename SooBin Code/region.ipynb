```python
# Dependencies and loading data

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

hp15 = 'data/2015.csv'
hp16 = 'data/2016.csv'
hp17 = 'data/2017.csv'
hp18 = 'data/2018.csv'
hp19 = 'data/2019.csv'
region = 'data/region.csv'

df15 = pd.read_csv(hp15)
df16 = pd.read_csv(hp16)
df17 = pd.read_csv(hp17)
df18 = pd.read_csv(hp18)
df19 = pd.read_csv(hp19)
dfRe = pd.read_csv(region)
```

```python
# Dropping unnecessary columns and renaming columns

df15 = df15[['Country', 'Happiness Rank']]
df16 = df16[['Country', 'Happiness Rank']]
df17 = df17[['Country', 'Happiness.Rank']]
df18 = df18[['Country', 'Rank']]
df19 = df19[['Country (region)', 'Ladder']]

df15 = df15.rename(columns = {'Happiness Rank': 'Rank 2015'})
df16 = df16.rename(columns = {'Happiness Rank': 'Rank 2016'})
df17 = df17.rename(columns = {'Happiness.Rank': 'Rank 2017'})
df18 = df18.rename(columns = {'Rank': 'Rank 2018'})
df19 = df19.rename(columns = {'Country (region)': 'Country', 'Ladder': 'Rank 2019'})
```

```python
# Merging

m1 = pd.merge(df15, df16, on = 'Country', how = 'outer')
m2 = pd.merge(m1, df17, on = 'Country', how = 'outer')
m3 = pd.merge(m2, df18, on = 'Country', how = 'outer')
m4 = pd.merge(m3, df19, on = 'Country', how = 'outer')
m5 = pd.merge(m4, dfRe, on = 'Country', how = 'left')
```

```python
# Saving for reference

m5.to_csv('data/master.csv')
```

```python
# Summary

df = m5.groupby('Region')
df = df.mean()
df
```

```python
# Transposing it for plot

df_T = df.T
df_T
```

```python
# Plotting regional graph

x_axis = ['2015', '2016', '2017', '2018', '2019']
ticks = np.arange(0, 5, 1)

df_T.plot(figsize = (7, 5))
plt.gca().invert_yaxis()

plt.title('Yearly Change of Ranking by Region', fontsize = 15)
plt.xticks(ticks = ticks, labels = x_axis)
plt.xlabel('Year', fontsize = 12)
plt.ylabel('Rank', fontsize = 12)
plt.legend(loc = 'best', bbox_to_anchor = (1, 0.7))
plt.grid()

plt.savefig('images/Ranking_region.png', bbox_inches = 'tight')

plt.show()
```

```python
# Historical happiness ranking of the query result

df_q = m5.set_index('Country')
canada = df_q.loc['Canada', ['Rank 2015', 'Rank 2016', 'Rank 2017', 'Rank 2018', 'Rank 2019']]
skorea = df_q.loc['South Korea', ['Rank 2015', 'Rank 2016', 'Rank 2017', 'Rank 2018', 'Rank 2019']]

plt.plot(x_axis, canada)
plt.plot(x_axis, skorea)
plt.xlim(0, len(x_axis) - 1)
plt.title(f'Historical Happiness Ranking of {canada.name} and {skorea.name}', fontsize = 15)
plt.xlabel('Year', fontsize = 12)
plt.ylabel('Rank', fontsize = 12)
plt.gca().invert_yaxis()
plt.legend(loc = 'best')
plt.grid()

plt.savefig('images/historical_ranking.png', bbox_inches = 'tight')

plt.show()
```

```python
# Plotting a boxplot

m5.boxplot('Rank 2019', by = 'Region', figsize = (7, 5))

plt.ylim(0, m5['Rank 2015'].count())
plt.xticks(rotation = 90)
plt.gca().invert_yaxis()

plt.xlabel('Region', fontsize = 12)
plt.ylabel('Rank', fontsize = 12)

plt.savefig('images/regional_boxplot.png')

plt.show()
```

```python
# Social support boxplot for 2018
data18 = pd.read_csv('data/2018.csv')
m6 = pd.merge(data18, dfRe, on = 'Country', how = 'left')

m6.boxplot('Social_Support', by = 'Region', figsize = (7, 5))

plt.xticks(rotation = 90)

plt.xlabel('Region', fontsize = 12)
plt.ylabel('Social_Support', fontsize = 12)

plt.savefig('images/social_boxplot.png')

plt.show()
```

```python
# Life expectancy boxplot for 2018

m6.boxplot('Healthy_Life_Expectancy', by = 'Region', figsize = (7, 5))

plt.xticks(rotation = 90)

plt.xlabel('Region', fontsize = 12)
plt.ylabel('Healthy Life Expectancy', fontsize = 12)

plt.savefig('images/life_boxplot.png')

plt.show()
```
