```python
import pandas as pd
```

```python
happy_df_2019 = pd.read_excel("../Data/world-happiness-report-2019-withregion.xlsx")
area_df = pd.read_csv("../Data/area.csv")
population_df = pd.read_csv("../Data/population.csv", names=["Rank","COUNTRY","Population","Year"])
birthrate_df = pd.read_excel("../Data/BirthRate2018.xlsx")
unemploy_df = pd.read_excel("../Data/Unemployment2017.xlsx")
obesity_df = pd.read_csv("../Data/obesity_rates.csv")
```

```python
#creating a DF out of just the area and population, to calculate population density
popdensity_df=pd.merge(area_df,population_df,on="COUNTRY",how="outer")
popdensity_df.head()
```

```python
#trim to only Country, SQKM, and Population
trimmed_popdens_df=pd.DataFrame({"Country":popdensity_df["COUNTRY"],"SQ KM":popdensity_df["(SQ KM)"],
                                            "Population":popdensity_df["Population"]})
trimmed_popdens_df.head()
```

```python
#dropna
trimmed_popdens_df=trimmed_popdens_df.dropna()
trimmed_popdens_df.head()
```

```python
trimmed_popdens_df["SQ KM"]=trimmed_popdens_df["SQ KM"].astype(float)
```

```python
#Make pop density calculation
#I had to use excel to change the formatting of the SQKM column to make this work. I tried using Regex but it didn't work well.
trimmed_popdens_df["Population Density"]=trimmed_popdens_df["Population"]/trimmed_popdens_df["SQ KM"]
trimmed_popdens_df.head()
```

```python
#renaming columns so that the happy_df_2019 pd and trimmed_popdens_df pd can be merged on "Country"
happy_df_2019=happy_df_2019.rename(columns={"Country (region)":"Country"})
happy_df_2019.head()
```

```python
#Making country names all lowercase in both datasets
happy_df_2019["Country"]=happy_df_2019["Country"].str.lower()
trimmed_popdens_df["Country"]=trimmed_popdens_df["Country"].str.lower()
```

```python
#merging the happiness data and the pop density data
merged_df=pd.merge(happy_df_2019,trimmed_popdens_df,on="Country",how="outer")
merged_df=merged_df.dropna()
merged_df.head()
```

```python
#Formatting Birthrate df to merge
trimmed_birthrate_df=pd.DataFrame({"Country":birthrate_df["COUNTRY"],
                                   "Birth Rate (Child per woman)":birthrate_df["(CHILDREN BORN/WOMAN)"]})
trimmed_birthrate_df["Country"]=trimmed_birthrate_df["Country"].str.lower()
trimmed_birthrate_df.head()
```

```python
#Formatting obesity df to merge
trimmed_obesity_df=pd.DataFrame({"Country":obesity_df["COUNTRY"],
                                   "Obesity Rate (%)":obesity_df["(%)"]})
trimmed_obesity_df["Country"]=trimmed_obesity_df["Country"].str.lower()
trimmed_obesity_df.head()
```

```python
#Formatting unemployment df to merge
trimmed_unemploy_df=pd.DataFrame({"Country":unemploy_df["Country"],
                                   "Unemployment Rate (%)":unemploy_df["Unemployment Rate (%)"]})
trimmed_unemploy_df["Country"]=trimmed_unemploy_df["Country"].str.lower()
trimmed_unemploy_df.head()
```

```python
#merge birthrate
merged_df=pd.merge(merged_df,trimmed_birthrate_df,on="Country",how="outer")
merged_df=merged_df.dropna()
merged_df.head()
```

```python
#merge obesity
merged_df=pd.merge(merged_df,trimmed_obesity_df,on="Country",how="outer")
merged_df=merged_df.dropna()
merged_df.head()
```

```python
#merge unemployment
merged_df=pd.merge(merged_df,trimmed_unemploy_df,on="Country",how="outer")
merged_df=merged_df.dropna()
merged_df.head()
```

```python
#checking we still have a reasonable number of countrys after all the dropnas
merged_df.count()
```

```python
#now the happiness rankings have been messed up - i.e. the original data would have 1 to 149 or so, and maybe country 20
# and 25 have been trimmed out, so it would go 1 .. 19, 21, 22, 23, 24, 26 etc. We want it to be 1 to 125
#for all criteria
#We are going to create a code that will sort a column, then assign 1 to 125 based on the order of that sorting in a
# new column, then sort by the next column, assign 1 to 125, etc.
```

```python
#rearrange df
clean_df = pd.DataFrame({"Country":merged_df["Country"],
                         "Region":merged_df["Region"],
                         "Ladder":merged_df["Ladder"],
                         "Social Support":merged_df["Social support"],
                         "Freedom":merged_df["Freedom"],
                         "Trust":merged_df["Corruption"],
                         "Generosity":merged_df["Generosity"],
                         "GDP per Capita":merged_df["Log of GDP\r\nper capita"],
                         "Healthy Life Expectancy":merged_df["Healthy life\r\nexpectancy"],
                         "Population Density":merged_df["Population Density"],
                         "Birth Rate":merged_df["Birth Rate (Child per woman)"],
                         "Obesity Rate (%)":merged_df["Obesity Rate (%)"],
                         "Unemployment Rate (%)":merged_df["Unemployment Rate (%)"],
                        })
clean_df.head()
```

```python
#getting a list of the 10 factor columns that need to be sorted then assigned a rank
columns = list(clean_df.columns)
columns.remove("Country")
columns.remove("Region")
columns
```

```python
#this for loop will sort a column, create a new column that is a ranking 1-126 for that sorting, then sort next column etc
clean_df["Obesity Rate (%)"]=clean_df["Obesity Rate (%)"] * (-1)
clean_df["Unemployment Rate (%)"]=clean_df["Unemployment Rate (%)"] * (-1)

for column in columns:
    
    test_list=[]
    for country in range (1,126):
        test_list.append(country)
        clean_df=clean_df.sort_values(column, axis = 0, ascending = True)
    clean_df[f"{column} rank"]= test_list

clean_df["Obesity Rate (%)"]=clean_df["Obesity Rate (%)"] * (-1)
clean_df["Unemployment Rate (%)"]=clean_df["Unemployment Rate (%)"] * (-1)
```

```python
#print to excel
clean_df.to_excel("../Data/MergedData.xlsx")
```
